```python
import pandas as pd
```

```python
bios = pd.read_csv('./bios.csv')

bios.head()
```

```python
bios.info()
```

```python
bios.loc[bios['height_cm'] > 215, ['name','height_cm'] ]
```

```python
bios.loc[(bios['height_cm'] > 215) & (bios['born_country']=='USA') , ['name','height_cm', 'born_country']]
```

```python
bios[bios['name'].str.startswith("Keith")]
```

```python
bios[bios['name'].str.contains("Keith" , case=False)]
```
