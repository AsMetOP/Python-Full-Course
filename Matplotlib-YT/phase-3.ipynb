Scatter 

```python
import matplotlib.pyplot as plt 

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,90]

plt.scatter(hours_studied,exam_scores , color='green', marker='^', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(not False)
plt.show()
```

```python
plt.scatter([1,2,3],[50,60,70] , color='green', marker='^', label='classA')
plt.scatter([4,5,6],[40,65,90] , color='red', marker='^', label='classB')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparison of two classes')
plt.legend()
plt.grid(not False)
plt.show()
```
