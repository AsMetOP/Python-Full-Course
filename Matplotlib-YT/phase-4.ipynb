Subplot 

```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [10,20,30,40]



plt.subplot(1,2,1)
plt.plot(x,y, marker='o')
plt.grid(True , linestyle=':')
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.grid(True , linestyle=':')
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,30,40]
fig, ax = plt.subplots(1,2, figsize= (10,5))
ax[0].plot(x,y)
ax[0].set_title('Line Chart')


ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of line and bar charts')
plt.tight_layout()
plt.show()
```
