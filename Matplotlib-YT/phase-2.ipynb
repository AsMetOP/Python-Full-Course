Plotting Bar Charts, Pie Charts, and Histrograms

Bar Charts --> Category Comparison, Data Analysis

Histograms --> Numerical Distribution , Data Insights (Data distributed )

Pie Charts --> Propotion illustration, Whole Representation

```python
import matplotlib.pyplot as plt 

product = ['A','B','C','D']
sales = [1000,2000,3000,4000]

plt.bar(product, sales, color='orange', label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()

plt.show
```

```python
regions = ['North' ,'South', 'East', 'West']
revenue = [3000,2000,4000,2500]

plt.pie(revenue ,labels=regions, autopct='%1.1f%%')
plt.title('Revenue Contribution By Region')
plt.show()
```

```python
scores = [45,23,24,76,69,65,90,40,10,14,87,98,57,45,47,48,82,19,17,67]

plt.hist(scores, bins=5, color='purple', edgecolor='red')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score distribution of students')
plt.show()
```
