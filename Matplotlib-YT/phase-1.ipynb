# Uses of Pyplot functions

```python
import matplotlib.pyplot as plt 
```

```python
x =[1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
day = ["Mon","Tu","Wed","Thr","Fri"]
sales = [100,40,29,15,5]

plt.plot(day,sales)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title("Bakery Sales this week! ")
plt.show()
```

```python
# py.plot(x,y, colour='colour_name', linestyle='line_style',linewidth = value,marker= 'marker_symbol',label= 'label_name')
months = [1,2,3,4]
sales = [10000, 1500, 6000, 3000]

plt.plot(months, sales, color='blue', linestyle='--',linewidth = 2,marker= 'o',label= '2025 sales data')
plt.legend(loc='upper right', fontsize=12)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title("Montly sales data report")
plt.grid(True , linestyle=':')
plt.xlim(1,3)
plt.ylim(0,5000)
plt.xticks([1,2,3,4], ['M1','M2','M3','M4'])
plt.show()
```
