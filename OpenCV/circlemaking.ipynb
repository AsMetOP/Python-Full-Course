Circle

```python
old_image = cv2.imread(r"images/Screenshot 2024-12-22 180436.png")
old_image = cv2.resize(old_image,(500,500))

#Thug
txt = cv2.putText(img= old_image,
text = "THUG",
org= (80,80),
fontFace=cv2.FONT_ITALIC,
fontScale= 1,
color=(240,0,0),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

#Mortal
txt1 = cv2.putText(img= old_image,
text = "MORTAL",
org= (300,120),
fontFace=cv2.FONT_ITALIC,
fontScale= 1,
color=(240,0,0),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

#Thug
new_image = cv2.circle(img= txt, center=(120, 140), radius= 44,color=(78,1,123), thickness=4 , lineType=4)
#Mortal
new_image = cv2.circle(img= txt1, center=(370, 190), radius= 44,color=(78,1,123), thickness=3 , lineType=4)

cv2.imshow("Founders",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Ellipse

```python
old_image = cv2.imread(r"images/Screenshot 2024-12-22 180436.png")
old_image = cv2.resize(old_image,(500,500))

#Thug
txt = cv2.putText(img= old_image,
text = "THUG",
org= (80,80),
fontFace=cv2.FONT_ITALIC,
fontScale= 1,
color=(240,0,0),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

#Mortal
txt1 = cv2.putText(img= old_image,
text = "MORTAL",
org= (300,120),
fontFace=cv2.FONT_ITALIC,
fontScale= 1,
color=(240,0,0),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

#Thug
new_image = cv2.ellipse(img= txt, center=(120, 140), axes= (28,48),angle=180 ,startAngle=0, endAngle= 360,color=(78,1,123), thickness=4 , lineType=4)
#Mortal
new_image = cv2.ellipse(img= txt1, center=(370, 190), axes= (26,48), angle=160,startAngle=0, endAngle=360 ,color=(78,1,123), thickness=3 , lineType=4)

cv2.imshow("Founders",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

PolyLine

```python
old_image = cv2.imread(r"images/Screenshot 2024-12-22 180436.png")
old_image = cv2.resize(old_image,(500,500))

#Mortal
txt1 = cv2.putText(img= old_image,
text = "MORTAL",
org= (300,120),
fontFace=cv2.FONT_ITALIC,
fontScale= 1,
color=(240,0,0),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)


#Mortal
new_image =cv2.polylines(img= txt1, pts= [np.array([[100,200],[200,100],[300,400],[600,100],[100,400]])],isClosed= True,color=(78,1,123), thickness=4 , lineType=4)

cv2.imshow("Founders",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
