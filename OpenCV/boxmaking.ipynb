```python
import cv2
```

```python
img = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-23 114732.png")

resize = cv2.resize(img,(400,400))
cv2.imshow("Founder", resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Line Draw 

new_image = cv2.line(img= resize, pt1=(180,60), pt2=(250,60), color=(0,250,2), thickness=4, lineType = 4)
cv2.imshow("Founder", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Rectange Draw
txt= cv2.putText(img= resize,
text = "ToxicMavi",
org= (150,50),
fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
fontScale= 1,
color=(0,0,255),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

new_img= cv2.rectangle(img= txt, pt1=(150,60), pt2=(300,190), color=(0,250,2), thickness=4, lineType=4)
cv2.imshow("Founder",new_img )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
