```python
#Multiple image using numpy in opencv

img = cv2.imread("images/Screenshot 2024-12-22 180436.png")
#Resize of image
resize = cv2.resize(img,(500,500))
h = np.hstack((resize,resize))
cv2.imshow("Founder of S8UL",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Multiple image using numpy in opencv

img = cv2.imread("images/Screenshot 2024-12-22 180436.png")
#Resize of image
resize = cv2.resize(img,(300,300))
v = np.vstack((resize,resize))
cv2.imshow("Founder of S8UL",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Multiple image using numpy in opencv

img = cv2.imread("images/Screenshot 2024-12-22 180436.png")
#Resize of image
resize = cv2.resize(img,(300,300))
h = np.hstack((resize,resize,resize))
v = np.vstack((h,h))
cv2.imshow("Founder of S8UL",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
list_name = os.listdir(r"D:\Python\OpenCV\images")
list_name


#Slideshow

for name in list_name:
    path = "D:\\Python\\OpenCV\\images"
    img_name = path + "\\" + name
    img = cv2.imread(img_name)
    img = cv2.resize(img,(500,600))
    cv2.imshow("Images",img)
    cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# cv2.IMREAD_COLOR --> 1 (No flag needed)

img = cv2.imread(r"images\Screenshot 2024-12-23 114732.png",1)
# print(img)  #rgb (0-255,0-255,0-255)
print(img.shape) #(774, 894, 3) - 3 here is the rgb
img = cv2.resize(img,(500,500))
cv2.imshow("Pictures",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# cv2.IMREAD_GRAYSCALE --> 0 (Black and White Image)

img = cv2.imread(r"images\Screenshot 2024-12-23 114732.png",0)
# print(img)  #rgb (0-255,0-255,0-255)
print(img.shape) #(774, 894, 3) - 3 here is the rgb
img = cv2.resize(img,(500,500))
cv2.imshow("Pictures",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
