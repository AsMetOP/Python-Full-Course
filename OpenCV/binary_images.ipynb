```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread(r"D:\Python\OpenCV\images\download-(6).png")
img2 = cv2.imread(r"D:\Python\OpenCV\images\Mw1iJ.png")
img1 = cv2.resize(img1,(200,200))
img2 = cv2.resize(img2,(200,200))

new = cv2.bitwise_and(img1,img2)
h = np.hstack((img1,img2,new))
cv2.imshow("BITWISE",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
