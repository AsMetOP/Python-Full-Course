```python
import cv2
import numpy as np
```

```python
img_get = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-23 114732.png")
resize = cv2.resize(img_get,(400,400))
cv2.imshow("Maavi",resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
txt = cv2.putText(img= resize,
text = "ToxicMavi",
org= (50,50),
fontFace=cv2.FONT_HERSHEY_DUPLEX,
fontScale= 3,
color=(0,0,255),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin= False)

cv2.imshow("Maavi",txt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
