```python
import cv2
```

```python
img1 = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-23 114732.png")
img2 = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-22 180436.png")

img1 = cv2.resize(img1,(500,500))
img2 = cv2.resize(img2,(500,500))

new_img = cv2.addWeighted(img1,0.5,img2,0.4,1)

cv2.imshow("Concatenated", new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-23 114732.png")
img2 = cv2.imread(r"D:\Python\OpenCV\images\Screenshot 2024-12-22 180436.png")

img1 = cv2.resize(img1,(500,500))
img2 = cv2.resize(img2,(500,500))

new_img = cv2.subtract(img2,img1)

cv2.imshow("SUBTRACTED", new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
