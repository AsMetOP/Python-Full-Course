```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])


print("-------Zomato Sales Analysis-------")
print ("\n Sales Data Shape :" ,sales_data.shape)
print ("\n Sales Data Type :" ,sales_data.dtype)
print ("\n Sales Data Dimensions :" ,sales_data.ndim)
```

```python
#Sample Data for First 3 Resturants
for i in range(0,3):
    print(sales_data[i])


#Another way :
print(sales_data[0:3])

#Ignores the first row
print(sales_data[:,1:])
```

```python
#Total sales per year
print(np.sum(sales_data[:,1:], axis=0))

#Minimum Sales Per Resturant
print(np.min(sales_data[:, 1:], axis=1))

#Minimum Sales Per Year
print(np.min(sales_data[:, 1:], axis=0))

#Maximum Sales Per Resturant
print(np.max(sales_data[:, 1:], axis=1))

#Maximum Sales Per Year
print(np.max(sales_data[:, 1:], axis=0))

#Average Sales per Year
print(np.mean(sales_data[:,1:], axis=0))

#Average Sales per Resturant
print(np.mean(sales_data[:,1:], axis=1))

```

```python
cumsum = np.cumsum(sales_data[:,1:],axis = 1)
print(cumsum)

plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales across all resturants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(False)
plt.show
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", np.add(vector1 , vector2))

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
